```python
import pandas as pd
import numpy as np
from pandas import json_normalize
import requests
```

```python
response = requests.get('https://api.themoviedb.org/3/tv/popular?api_key=f968ca10727e6a94e284e3e08828e918&language=en-US&with_original_language=hi&region=IN&page=1')
```

```python
response.json()['results']
```

```python
df = pd.DataFrame()
for i in range(1,64):
    response = requests.get('https://api.themoviedb.org/3/tv/popular?api_key=f968ca10727e6a94e284e3e08828e918&language=en-US&with_original_language=hi&region=IN&page={}'.format(i))
    temp_df = pd.DataFrame(response.json()['results'])
    df=pd.concat([df,temp_df], ignore_index=True)
```

```python
df
```

```python
credits = pd.DataFrame(columns=['cast','crew'])
for i in range(0,1258):
    response = requests.get('https://api.themoviedb.org/3/tv/{}/credits?api_key=f968ca10727e6a94e284e3e08828e918&language=en-US'.format(df['id'].iloc[i]))
    array = response.json()
    temp_cast = json_normalize(array, 'cast')
    temp_crew = json_normalize(array,'crew')
    d = temp_cast.to_dict(orient='index')
    e = temp_crew.to_dict(orient='index')
    list = []
    list2 = []
    for i in d.values():
        list.append(i)
    for i in e.values():
        list2.append(i)
    credits=credits.append({'cast':list,'crew':list2},ignore_index=True)
```

```python
credits
```

```python
keywords=pd.DataFrame(columns=['keywords'])
for i in range(0,1258):
    response=requests.get('https://api.themoviedb.org/3/tv/{}/keywords?api_key=f968ca10727e6a94e284e3e08828e918'.format(df['id'].iloc[i]))
    array = response.json()
    temp_keywords=json_normalize(array,'results')
    d=temp_keywords.to_dict(orient='index')
    list=[]
    a=''
    for i in d.values():
        a=i
        a.pop('id')
        for i in a.values():
            list.append(i)
    keywords=keywords.append({'keywords':list},ignore_index=True)
```

```python
keywords
```

```python
genres=pd.DataFrame(columns=['genres'])
for i in range(0,5871):
    response=requests.get('https://api.themoviedb.org/3/tv/{}?api_key=f968ca10727e6a94e284e3e08828e918&language=en-US'.format(df['id'].iloc[i]))
    array = response.json()
    temp_genres=json_normalize(array, 'genres')
    d=temp_genres.to_dict(orient='index')
    list=[]
    for i in d.values():
        list.append(i)
    genres=genres.append({'genres':list},ignore_index=True)
```

```python
genres
```

```python
df.drop(labels='genre_ids', axis="columns", inplace=True)
```

```python
df = pd.concat([df, genres], axis=1, join='inner')
df = pd.concat([df, keywords], axis=1, join='inner')
df = pd.concat([df, credits], axis=1, join='inner')
```

```python
df.rename(columns = {'id':'movie_id'}, inplace = True)
```

```python
df.rename(columns = {'movie_id':'show_id'}, inplace = True)
```
